```python
import pandas as pd
import numpy as np
import mlxtend
from mlxtend.preprocessing import TransactionEncoder

dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns = te.columns_)
df.head()

```

```python
df.info()
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, use_colnames = True)
apriori(df, use_colnames = True)
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, min_threshold=0.7, metric='confidence')
```

We can notice that couples items are oftenly sold together, such as Skirt and Pant, or Scarf and Pants,
It could be a good marketing strategy to put these items together.

```python
import csv 

items = []

with open('Market_Basket_Optimisation.csv') as csvfile:    
	csvReader = csv.reader(csvfile)    
	for row in csvReader:        
		items.append(row)     
        
te_array = te.fit(items).transform(items)
df = pd.DataFrame(te_array, columns = te.columns_)
df.head()

```
